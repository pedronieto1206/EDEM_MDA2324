```python
# read json to data frame
df_University=pd.read_csv('/gdrive/MyDrive/EDEM/archive/World University Rankings 2023.csv')
df_University.head(1)
```

```python
df_University=df_University[df_University['Research Score'].notnull()]
df_University=df_University[df_University['No of student'].notnull()]
df_University=df_University[df_University['Female:Male Ratio'].notnull()]
```

```python
df_University.dtypes
```

*Change data types*

```python
# Extraer los tres primeros caracteres y crear una nueva columna
#df_University['No of student int'] = df_University['No of student'].str[:2]
#df_University['No of student int2'] = df_University['No of student'].str[3:-1]
#df_University['University Rank'] = df_University['University Rank'].astype(int)
df_University['No of student per staff 2']= df_University['No of student per staff'].astype(float)
df_University['Female:Male Ratio']= df_University['Female:Male Ratio'].astype(str)

#cambiar la , a un .
df_University['No of student']= df_University['No of student'].astype(str)
#df_University['No of student']= df_University['No of student'].astype(int)
df_University['No of student']= df_University['No of student'].replace(',', '.')

#df_University['No of student']= df_University['No of student'].replace(',', '.').astype(int)
```

```python
def to_int(val):
    """ Reconoce valores numericos y los transforma a enteros.
    """
    try:
        value = int(float(val))
    except ValueError:
        value = ""
    return value
```

```python
df_University['No of student'] = df_University['No of student'].map(to_int)
```
